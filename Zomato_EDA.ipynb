```python
# lets do the analysis of ratings section
ratings = df_final.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'rating count'})
ratings
```

```python
ratings.head()
```

```python
import matplotlib
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12, 6)
sns.barplot(x = 'Aggregate rating', y = 'rating count', data=ratings)
```

```python
sns.barplot(x = 'Aggregate rating', y = 'rating count',  hue='Rating color',data=ratings, palette=['blue','red','orange','yellow','green','darkgreen'])
```

```python
#conclusion:- not rated count is very high , max ratings are between range 3.1 and 3.2
```

```python
sns.countplot(x = 'Rating color', data=ratings, palette=['blue','red','orange','yellow','green','darkgreen'] )
```
